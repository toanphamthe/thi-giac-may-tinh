```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'C:/Users/DevG/OneDrive/Documents/Thi Giac May Tinh/Image/01/01.jpg'
im = Image.open(image_path)
im_array = np.array(im)
print(im_array.shape)

im_array_2 = np.array(im.convert('L'), 'f')
print(im_array_2.shape)
print(im_array_2.dtype)
```

```python
image = Image.open(image_path)
image_array = np.array(image.convert("L"))

inverted_image = 255 - image_array
squared_image = image_array ** 2
clamped_image = np.clip(image_array, 100, 200)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(image_array, cmap="gray")
axs[0, 0].set_title('Hnh nh gc')

axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Hnh nh  bin i')

axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Hnh nh  bin i')

axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Hnh nh  bin i')

plt.show()
```
