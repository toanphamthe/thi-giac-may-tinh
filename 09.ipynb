```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'C:/Users/DevG/OneDrive/Documents/Thi Giac May Tinh/Image/01/01.jpg'
im = Image.open(image_path)
plt.imshow(im)
```

```python
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
